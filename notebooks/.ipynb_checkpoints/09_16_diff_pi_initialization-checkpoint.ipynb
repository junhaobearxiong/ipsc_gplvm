For this experiment, look at the effect of different initialization of pi on split-GPLVM. Initializations of pi's are:
* True labels, fixed across training
* True labels, learn across training
* Random labels, learn across training
* Uniformly random, learn across training

All models are optimized w/ natgrad + adam. Z's are learned and initialized w/ the same values across models.

Result
* Regardless of initialization, model always prefers uniform mixture weights, which also results in higher ELBO. The lower ELBO w/ mixture model partly accounted for by KL(q(pi) | p(pi)). So maybe mixture only slightly increase the likelihood, but is penalized by KL

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import gpflow
import pandas as pd

from gpflow import set_trainable
from sklearn.decomposition import PCA

from utils import *
```

```python
np.random.seed(1)

num_data = 200
dim = 5

xmin = 0
xmax = 10
break_pt = 3
k1 = 0.0
k21 = 0.3
k22 = -0.3
c1 = 0
sigma = 0.75
noise = 0.0
lengthscales = [0.2, 5, 1]

x = np.linspace(xmin, xmax, int(num_data/2), dtype=default_float())
X = branch_simulation(x, break_pt, k1, k21, k22, c1, sigma)
labels = np.repeat([0, 1], int(num_data)/2)
labels_dummy = pd.get_dummies(labels).values
Y = gen_obs(X, True, labels, dim, lengthscales, noise)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=labels, ax=axs[0])
sns.scatterplot(x=Y[:, 0], y=Y[:, 1], hue=labels, ax=axs[1])
sns.scatterplot(x=Y[:, 0], y=Y[:, 1], hue=X[:, 0], ax=axs[2])

axs[0].set_title('Latent space')
axs[0].set_xlabel('x1')
axs[0].set_ylabel('x2')

axs[1].set_title('Observed space: 2D, color by labels')
axs[1].set_xlabel('y1')
axs[1].set_ylabel('y2')

axs[2].set_title('Observed space: 2D, color by x1')
axs[2].set_xlabel('y1')
axs[2].set_ylabel('y2')
```

```python
# fix inducing inputs initialization

np.random.seed(1)
M = 20
X_pca = PCA(n_components=2).fit_transform(Y)
Zp = np.random.permutation(X_pca[:, :1])[:M]
Zs = np.random.permutation(X_pca[:, :1])[:M]
```

```python
# true labels, fixed
m1 = init_split_gplvm(
    Y=Y,
    split_space=True,
    Qp=1,
    Qs=1,
    K=2,
    Zp=Zp,
    Zs=Zs,
    pi_init=labels_dummy + 1e-4
)
set_trainable(m1.pi, False)

# true labels, learn
m2 = init_split_gplvm(
    Y=Y,
    split_space=True,
    Qp=1,
    Qs=1,
    K=2,
    Zp=Zp,
    Zs=Zs,
    pi_init=labels_dummy + 1e-4
)

# random labels, learn
m3 = init_split_gplvm(
    Y=Y,
    split_space=True,
    Qp=1,
    Qs=1,
    K=2,
    Zp=Zp,
    Zs=Zs,
    pi_init=np.random.permutation(labels_dummy) + 1e-4
)

# uniform initialization
m4 = init_split_gplvm(
    Y=Y,
    split_space=True,
    Qp=1,
    Qs=1,
    K=2,
    Zp=Zp,
    Zs=Zs
)
```

```python
%%time
elbo1, pi1 = train_natgrad_adam(m1, 3000, 500)
```

```python
%%time
elbo2, pi2 = train_natgrad_adam(m2, 3000, log_freq=100)
```

```python
%%time
elbo3, pi3 = train_natgrad_adam(m3, 3000, log_freq=100)
```

```python
%%time
elbo4, pi4 = train_natgrad_adam(m4, 3000, log_freq=100)
```

```python
# save the max of assignment probabilities among all samples 
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

pis = [pi1, pi2, pi3, pi4]
names = ['true labels, fixed', 'true labels, learn', 'random labels, learn', 'uniform, learn']

for i, logpi in enumerate(pis):
    maxpi = []
    for pi in logpi:
        maxpi.append(pi.numpy().max())
    axs[0].plot(maxpi, label=names[i])
    
axs[0].set_xlabel('iterations')
axs[0].set_ylabel('max pi')
axs[0].legend()

elbos = [elbo1, elbo2, elbo3, elbo4]
for i, elbo in enumerate(elbos):
    axs[1].plot(elbo, label=names[i])
axs[1].set_xlabel('iterations')
axs[1].set_ylabel('ELBO')
axs[1].legend()
```
