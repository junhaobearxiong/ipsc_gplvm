```python
def breakpoint_linear(x, ts, k1, k2, c1, sigma):
    '''Function representing a step-wise linear curve with one
    breakpoint located at ts.
    '''
    func = np.piecewise(x, [x < ts], [lambda x: k1 * x + c1,
                                      lambda x: k2 * x + (k1 - k2) * ts + c1])
    return func + np.random.normal(func, sigma)
```

```python
# X ([num_data, 2]), where X[:,0] is "time", X[:,1] is the values of two branching processes 
def branch_simulation(x, break_pt, k1, k21, k22, c1, sigma):
    func1 = breakpoint_linear(x, break_pt, k1, k21, c1, sigma)
    func2 = breakpoint_linear(x, break_pt, k1, k22, c1, sigma)
    data1 = np.stack([x, func1], axis=1)
    data2 = np.stack([x, func2], axis=1)
    X = np.concatenate([data1, data2], axis=0)
    return X #(X - X.mean()) / X.std()
```

```python
# generate high dimensional observation Y ([num_data, dim])
# by mapping X to dim-dimensional space through a GP w/ kernel function
def gen_obs(X, labels, dim, lengthscales, obs_noise=1):
    N = X.shape[0]
    Y = np.zeros([N, dim])
    
    k1 = gpflow.kernels.SquaredExponential(lengthscales=lengthscales[0])
    k2 = gpflow.kernels.SquaredExponential(lengthscales=lengthscales[1])
    k3 = gpflow.kernels.SquaredExponential(lengthscales=lengthscales[2])
    # k4 = gpflow.kernels.SquaredExponential(lengthscales=lengthscales[3])
    
    label_counts = np.unique(labels, return_counts=True)[1]
    
    # y = fs(x1) + fk(x2), where k depends on the label of x
    # each fk has a different lengthscale
    # fs(x1)
    Y = np.random.default_rng().multivariate_normal(np.zeros(N), k1(X[:, 0][:, np.newaxis]), dim).T
    # fk(x2), k=1
    Y[labels == 0, :] += np.random.default_rng().multivariate_normal(np.zeros(label_counts[0]), 
                                                                   k2(X[labels==0, 1][:, np.newaxis]), dim).T
    # fk(x2), k=2
    Y[labels == 1, :] += np.random.default_rng().multivariate_normal(np.zeros(label_counts[1]),
                                                                   k3(X[labels==1, 1][:, np.newaxis]), dim).T
    # fk(x2), k=3
    #Y[labels == 2, :] += np.random.default_rng().multivariate_normal(np.zeros(label_counts[2]),
    #                                                               k4(X[labels==2, 1][:, np.newaxis]), dim).T
    return Y #(Y - Y.mean()) / Y.std()
```

```python
def plot_kernel_samples(k, ax, xmin=-3, xmax=3):
    xx = np.linspace(xmin, xmax, 100)[:, None]
    K = k(xx)
    ax.plot(xx, np.random.multivariate_normal(np.zeros(100), K, 3).T)
```

```python
np.random.seed(1)

num_data = 200
dim = 5

xmin = 0
xmax = 10
break_pt = 3
k1 = 0.0
k21 = 0.3
k22 = -0.3
c1 = 0
sigma = 0.75
noise = 0.2
#lengthscales = [1.0, 0.2, 5]
lengthscales = [1.0, 0.2, 5]


x = np.linspace(xmin, xmax, int(num_data/2), dtype=default_float())
X = branch_simulation(x, break_pt, k1, k21, k22, c1, sigma)
#labels = np.concatenate([np.where(x < break_pt, 0, 1), np.where(x < break_pt, 0, 2)])
labels = np.repeat([0, 1], int(num_data)/2)
Y = gen_obs(X, labels, dim, lengthscales, noise)
labels_dummy = pd.get_dummies(labels).values
```

## GPLVM

```python
def step_callback(step, variables, values):
    print('step {}'.format(step))

def train_scipy(m, maxiter):
    opt = gpflow.optimizers.Scipy()
    _ = opt.minimize(
        m.training_loss,
        method="BFGS",
        variables=m.trainable_variables,
        options=dict(maxiter=ci_niter(maxiter), disp=True),
        #step_callback=step_callback,
        compile=True
    )
```

```python
def init_gplvm(Y, latent_dim, num_inducing, kernel):
    num_data = Y.shape[0]  # number of data points

    X_mean_init = tf.constant(PCA(n_components=latent_dim).fit_transform(Y), dtype=default_float())
    X_var_init = tf.ones((num_data, latent_dim), dtype=default_float())

    inducing_variable = tf.convert_to_tensor(np.random.permutation(X_mean_init.numpy())[:num_inducing], dtype=default_float())
    #inducing_variable = tf.convert_to_tensor(np.linspace(np.min(X_mean_init, axis=0), np.max(X_mean_init, axis=0), num_inducing), dtype=default_float())

    m = gpflow.models.BayesianGPLVM(
        Y,
        X_data_mean=X_mean_init,
        X_data_var=X_var_init,
        kernel=kernel,
        inducing_variable=inducing_variable,
    )
    return m
```

```python
np.random.seed(2)  # for reproducibility

M = 20

kernel1 = gpflow.kernels.SquaredExponential(lengthscales=[1.0])
kernel2 = gpflow.kernels.SquaredExponential(lengthscales=[1.0] * 2)

gplvm1 = init_gplvm(Y, latent_dim=1, num_inducing=M, kernel=kernel1)
gplvm2 = init_gplvm(Y, latent_dim=2, num_inducing=M, kernel=kernel2)
```

```python
%%time
train_scipy(gplvm1, maxiter=2000)
```

```python
%%time
train_scipy(gplvm2, maxiter=5000)
```

## Split-GPLVM

```python
def init_split_gplvm(Y, Qs, Qp, M, K, pi_init):
    N = Y.shape[0]  # number of data points

    X_pca = PCA(n_components=Qs+Qp).fit_transform(Y)
    Xs_mean_init = tf.constant(X_pca[:, :Qs], dtype=default_float())
    Xp_mean_init = tf.constant(X_pca[:, :Qs], dtype=default_float())
    Xs_var_init = tf.ones((N, Qs), dtype=default_float())
    Xp_var_init = tf.ones((N, Qp), dtype=default_float())

    Zs = tf.convert_to_tensor(np.random.permutation(Xs_mean_init.numpy())[:M], dtype=default_float())
    Zp = tf.convert_to_tensor(np.random.permutation(Xp_mean_init.numpy())[:M], dtype=default_float())

    kernel_s = gpflow.kernels.SquaredExponential(lengthscales=tf.convert_to_tensor([1.0] * Qs, dtype=default_float()))
    kernel_K = [gpflow.kernels.SquaredExponential(lengthscales=tf.convert_to_tensor([1.0] * Qp, dtype=default_float())) for k in range(K)]

    split_gplvm = SplitGPLVM(
        Y,
        Xs_mean=Xs_mean_init,
        Xs_var=Xs_var_init,
        Xp_mean=Xp_mean_init,
        Xp_var=Xp_var_init,
        pi=pi_init,
        kernel_s=kernel_s,
        kernel_K=kernel_K,
        Zs=Zs,
        Zp=Zp
    )

    set_trainable(split_gplvm.q_mu, False)
    set_trainable(split_gplvm.q_sqrt, False)
    set_trainable(split_gplvm.Zs, False)
    set_trainable(split_gplvm.Zp, False)
    
    return split_gplvm
```

```python
@tf.function
def optimization_step(model: SplitGPLVM, variational_params: List[Tuple[gpflow.Parameter, gpflow.Parameter]],
                     natgrad_opt: gpflow.optimizers.NaturalGradient,
                     adam_opt: tf.optimizers.Adam):
    natgrad_opt.minimize(model.training_loss, var_list=variational_params)
    adam_opt.minimize(model.training_loss, var_list=model.trainable_variables)
    # return model.elbo()
    return (model.elbo(), model.pi)


def train_natgrad_adam(model: SplitGPLVM, num_iterations: int, log_freq: int):
    natgrad_opt = NaturalGradient(gamma=1.0)
    adam_opt = tf.optimizers.Adam(learning_rate=0.01)
    variational_params = [(model.q_mu, model.q_sqrt)]

    logf = []
    logpi = []
    tol = 1e-4

    tf.print('initial elbo {:.4f}'.format(model.elbo()))

    for step in range(num_iterations):
        start_time = time.time()
        #elbo = optimization_step(model, variational_params, natgrad_opt, adam_opt)
        elbo, pi = optimization_step(model, variational_params, natgrad_opt, adam_opt)
        logf.append(elbo)
        logpi.append(pi)

        if step > 0 and np.abs(elbo - logf[-2]) < tol:
            tf.print('converge at iteration {} elbo {:.4f}'.format(step+1, elbo))
            break
        if (step + 1)  % log_freq == 0:
            tf.print('iteration {} elbo {:.4f}, took {:.4f}s'.format(step+1, elbo, time.time()-start_time))
            
    return (logf, logpi)
```

```python
%%time
split1 = init_split_gplvm(Y, Qs=1, Qp=1, M=20, K=2, pi_init=tf.ones((200, 2), dtype=default_float()))
log1 = train_natgrad_adam(split1, 2000, 100)
```

```python
%%time
split2 = init_split_gplvm(Y, Qs=1, Qp=1, M=20, K=2, 
                          pi_init=tf.constant(labels_dummy + 1e-4, dtype=default_float()))
#set_trainable(split2.pi, False)
log2, logpi2 = train_natgrad_adam(split2, 2000, 100)
```

```python
%%time
labels_copy = labels_dummy.copy()
np.random.shuffle(labels_copy)
split3 = init_split_gplvm(Y, Qs=1, Qp=1, M=20, K=2, pi_init=tf.constant(labels_copy + 1e-4, dtype=default_float()))
log3, logpi3 = train_natgrad_adam(split3, 2000, 100)
```

```python
%%time
split4 = init_split_gplvm(Y, Qs=1, Qp=1, M=20, K=2, 
                          pi_init=tf.constant(labels_dummy + 1e-4, dtype=default_float()))
set_trainable(split4.pi, False)
log4, _ = train_natgrad_adam(split4, 2000, 100)
```

### 09/15
For this set of experiments, we always have both shared and private space, and the simulation also Y based on fs and fk separately. Main conclusion is the categories did not help increase the likelihood of the data, but penalized by KL, so model always prefers uniform assignment. And the model did not recover lengthscales or informative latent embeddings (the difference between runs are probably due to different initializations of Z, since we are not learning them here)

```python
# save the max of assignment probabilities among all samples 
maxpi2 = []
maxpi3 = []
for i in range(len(logpi2)):
    maxpi2.append(logpi2[i].numpy().max())
    maxpi3.append(logpi3[i].numpy().max())
```

```python
plt.figure(figsize=(8, 6))
x = np.arange(1, 2001)
plt.plot(x, maxpi2, label='init w/ true labels')
plt.plot(x, maxpi3, label='init w/ random labels')
plt.xlabel('iterations')
plt.ylabel('max pi')
plt.legend()
```

```python
plt.figure(figsize=(8, 6))
x = np.arange(1, 2001)
plt.plot(x, log1, label='uniform')
plt.plot(x, log2, label='true labels, learn pi')
plt.plot(x, log3, label='random labels')
plt.plot(x, log4, label='true labels, fix pi')
plt.legend()
plt.title('different initialization of pi')
plt.xlabel('iterations')
plt.ylabel('ELBO')
```

Basically, assignments to different private space doesn't help explaining the data, i.e. doesn't increase likelihood, but is being penalized by KL, so the model prefers uniform assignment probabilities
